UnionWith:

```polyglot-notebook
HashSet<string> set1 = new HashSet<string>() { "a", "b", "c", "e" };
HashSet<string> set2 = new HashSet<string>() { "a", "c", "d", "e"};

set1.UnionWith(set2);

Console.WriteLine("Union of Set1 and Set2:");
foreach (string item in set1)
{
    Console.WriteLine(item);
}
```

IntersectWith:

```polyglot-notebook
HashSet<string> set1 = new HashSet<string>() { "a", "b", "c", "e" };
HashSet<string> set2 = new HashSet<string>() { "a", "c", "d", "e"};

set1.IntersectWith(set2);

Console.WriteLine("\nSet1 after intersecting with Set2:");
foreach (string item in set1)
{
    Console.WriteLine(item);
}
```

ExceptWith:

```polyglot-notebook


HashSet<string> set1 = new HashSet<string>() { "a", "b", "c", "e" };
HashSet<string> set2 = new HashSet<string>() { "a", "c", "d", "e"};

set1.ExceptWith(set2);

Console.WriteLine("\nSet1 after removing elements in Set2:");
foreach (string item in set1)
{
    Console.WriteLine(item);
}
```

## Queue\<T> Example

```polyglot-notebook
Queue<string> queue = new Queue<string>();
 
// Adding elements to the queue
queue.Enqueue("Apple");
queue.Enqueue("Banana");
queue.Enqueue("Guava");
queue.Enqueue("Orange");
 
// Check the number of elements in the queue
int count = queue.Count;
Console.WriteLine($" There are {count} elements in the queue.");             
 
// Print all the elements from the queue
int index = 0;
foreach(var element in queue)
{
    Console.WriteLine($" Element at index {index}: {element}");
    index++;
}
 
// Remove and return the first element in the queue
string front = queue.Dequeue();
Console.WriteLine($" Removing the element in front of the Queue: {front}");

// Check if the queue contains a specific element
bool containsApple = queue.Contains("Apple");
Console.WriteLine($" Does the queue contain apple? {containsApple}");

Console.WriteLine($" Peeked at: {queue.Peek()}");
```

### Stack\<T> Example

```polyglot-notebook
Stack<string> myStack = new Stack<string>();
myStack.Push("First Entry");
myStack.Push("Second Entry");
myStack.Push("Third Entry");

Console.WriteLine("Peeking at: " + myStack.Peek());

myStack.Pop();
Console.WriteLine("\nAfter popping:");
foreach(var item in myStack)
{
    Console.WriteLine($"{item}");
}

Console.WriteLine(myStack.Contains("Third Entry"));
```

Covariance example:

```polyglot-notebook
class Animal {}
class Dog : Animal {}

// Covariance example with interfaces
interface IAnimalProvider<out T>
{
    T GetAnimal();
}

class DogProvider : IAnimalProvider<Dog>
{
    public Dog GetAnimal()
    {
        return new Dog();
    }
}

// Covariance usage
IAnimalProvider<Animal> animalProvider = new DogProvider();
Animal animal = animalProvider.GetAnimal(); // Covariance allows returning Dog when Animal is expected
```

Contravariance example:

```polyglot-notebook
// Contravariance example with delegates
delegate void AnimalAction<in T>(T animal);

void FeedAnimal(Animal animal) { Console.WriteLine("Feeding animal"); }

// Contravariance usage
AnimalAction<Dog> dogAction = FeedAnimal;

dogAction(new Dog()); // Contravariance allows passing Dog when Animal is expected
```
